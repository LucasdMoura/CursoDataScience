```python
# Nesse trecho apresento o código importando a biblioteca pandas para fazer leitura de um arquivo de dados ".csv"
import pandas as pd

uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula4.1/movies.csv"
filmes = pd.read_csv(uri) #data frame
filmes.head() # head limita a exibição a somente os cinco primeiros da lista
```

```python
# Aqui traduzo o cabeçalho da minha tabela de informações.
filmes.columns = ["filmesId", "titulo", "generos" ]
filmes.head() # head limita a exibição a somente os cinco primeiros da lista
```

```python
# Agora foi inserida as avaliações acerca dos filmes na lista anterior em outro arquivo ".csv"
uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula1.2/ratings.csv"
notas = pd.read_csv(uri)
notas.head() # head limita a exibição a somente os cinco primeiros da lista
```

```python
# Tradução do cabeçalho da tabela
notas.columns = ["usuarioId", "filmesId", "nota", "momento"]
notas.head()
```

```python
# Avaliações únicas feitas por usuários
notas["nota"].unique()
```

```python
notas["nota"].mean
```

```python
# Estatísticas da lista estendida que lista contador, média, desvio padrão, minimo, 25%, 50%, 75%, maximo.
notas.describe()
```

```python
 # Gráfico de notas de filmes na lista extendida
notas.nota.plot(kind='hist')
```

```python
# Media de Avaliação pelo agrupamento "filmesID"
medias_por_filme = notas.groupby ("filmesId").mean().nota
medias_por_filme.head()
```

```python
# Gráfico de medias de notas na lista estendida
medias_por_filme.plot(kind='hist')
```
